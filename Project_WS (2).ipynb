```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

Brand = []
uniqueid = []
Car_Availability = []
Range = []
Efficiency = []
Weight = []

speed_0_100 = []
Battery = []
fast_charge = []
Towing = []
Cargo_Vol = []
Price_Range = []

Germany_Price = []
Netherlands_Price = []
United_Kingdom_Price = []

Charging_capability = []
Seats = []
Safety_Rating = []

for page in range(0, 10):

    start_time = time.time()

    URL = f"https://ev-database.org/#group=vehicle-group&rs-pr=10000_100000&rs-er=0_1000&rs-ld=0_1000&rs-ac=2_23&rs-dcfc=0_400&rs-ub=10_200&rs-tw=0_3000&rs-ef=100_350&rs-sa=-1_5&rs-w=1000_3500&rs-c=0_5000&rs-y=2010_2030&s=1&p={page}"
    
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")

    for car in soup.find_all('div', class_='item-data'):

        tag = car.find('a', class_='title')

        if tag and tag.has_attr('href'):
            href = tag['href']
            Brand.append(href.split('/')[-1].split('-')[0])
            uniqueid.append(href)
        else:
            Brand.append(np.nan)
            uniqueid.append(np.nan)

        if car.find('span', class_='erange_real'):
            Range.append(car.find('span', class_='erange_real').text.strip())
        else:
            Range.append(np.nan)

        if car.find('span', class_='efficiency'):
            Efficiency.append(car.find('span', class_='efficiency').text.strip())
        else:
            Efficiency.append(np.nan)

        if car.find('span', class_='acceleration_p'):
            speed_0_100.append(car.find('span', class_='acceleration_p').text.strip())
        else:
            speed_0_100.append(np.nan)

        if car.find('span', class_='battery_p'):
            Battery.append(car.find('span', class_='battery_p').text.strip())
        else:
            Battery.append(np.nan)

        if car.find('span', class_='fastcharge_speed_print'):
            fast_charge.append(car.find('span', class_='fastcharge_speed_print').text.strip())
        else:
            fast_charge.append(np.nan)

        if car.find('span', class_='cargo'):
            Cargo_Vol.append(car.find('span', class_='cargo').text.strip())
        else:
            Cargo_Vol.append(np.nan)

        if car.find('span', class_='priceperrange_p'):
            Price_Range.append(car.find('span', class_='priceperrange_p').text.strip())
        else:
            Price_Range.append(np.nan)

        if car.find('span', class_='country_de'):
            Germany_Price.append(car.find('span', class_='country_de').text.strip())
        else:
            Germany_Price.append(np.nan)

        if car.find('span', attrs={'data-tooltip': 'Market Segment'}):
            Charging_capability.append(
                car.find('span', attrs={'data-tooltip': 'Market Segment'}).text.strip()
            )
        else:
            Charging_capability.append(np.nan)
            
    print(f"Page {page + 1} scraped in {round(time.time() - start_time, 2)} seconds")


df = pd.DataFrame({
    "Brand": Brand,
    "UniqueID": uniqueid,
    "Range": Range,
    "Efficiency": Efficiency,
    "0-100": speed_0_100,
    "Battery": Battery,
    "FastCharge": fast_charge,
    "Cargo": Cargo_Vol,
    "Price_per_Range": Price_Range,
    "Germany_Price": Germany_Price,
    "Segment": Charging_capability,

})

print("\nTotal Rows:", len(df))
df
```

```python
df = df.rename(columns={
    "Range": "Range (Km)",
    "Efficiency": "Efficiency (Wh/km)",
    "Weight": "Weight (kg)",
    "0-100": "Acceleration (0-100)",
    "Battery": "Battery (kWh)",
    "FastCharge": "FastCharge (kW)",
    "Towing": "Towing (kg)",
    "Cargo": "Cargo (L)",
    "Price_per_Range": "Price_per_Range (/km)",
    "Germany_Price" : "Germany_Price(€)"
})
```

```python
import re

df["UniqueID"] = df["UniqueID"].astype('str').str.extract(r"(\d+)").astype('int')
```

```python
df["Range (Km)"] = df["Range (Km)"].astype('str').str.extract(r"(\d+)").astype('int')
```

```python
df["Efficiency (Wh/km)"] = df["Efficiency (Wh/km)"].astype('str').str.extract(r"(\d+)").astype('int')
```

```python
df["Acceleration (0-100)"] = (
    df["Acceleration (0-100)"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")
    .astype(float)
)
```

```python
df["Battery (kWh)"] = (
    df["Battery (kWh)"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")
    .astype(float)
)
```

```python
df["FastCharge (kW)"] = (
    df["FastCharge (kW)"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)
```

```python
df["Cargo (L)"] = (
    df["Cargo (L)"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)
)
```

```python

df["Germany_Price(€)"] = (
    df["Germany_Price(€)"].astype(str)
    .str.extract(r"(\d[\d,]*)")[0]
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
segment_mapping = {
    "A": "A - Mini",
    "B": "B - Compact",
    "C": "C - Medium",
    "D": "D - Large",
    "E": "E - Executive",
    "F": "F - Luxury",
    "N": "N - Passenger Van",
    "S": "S - Sports"
}

df["Segment"] = (
    df["Segment"]
    .astype(str)
    .str.strip()
    .str[0]              # Take first letter (A/B/C/D/E/F/N/S)
    .map(segment_mapping)
)
```

```python
df["Price_per_Range (/km)"] = (
    df["Price_per_Range (/km)"]
    .astype(str)
    .str.replace("€", "", regex=False)
    .str.replace("/km", "", regex=False)
    .str.strip()
    .astype(float)
)
```
